Write a "counter" class that can be incremented up to a specified maximum value, will print an error if an attempt is made to increment beyond that value, and allows reseting the counter.

```python
class Counter:
    def __init__(self, max_value):
        self.max_value = max_value
        self.value = 0

    def increment(self):
        if self.value < self.max_value:
            self.value += 1
        else:
            print("Error: Exceeded the maximum value.")

    def reset(self):
        self.value = 0

    def get_value(self):
        return self.value
```

```python
counter = Counter(3)

counter.increment()
print("Value:" ,counter.get_value())   

counter.increment()
print("Value:" ,counter.get_value())    

counter.increment()
print("Value:" ,counter.get_value())   

counter.increment()         
print("Value:" ,counter.get_value())   

counter.reset()
print("Value:" ,counter.get_value())  
```

```python
class Counter:
    
    def __init__(self, max_value):
        self.__max_value = max_value
        self.__value = 0

    def increment(self):
        if self.__value < self.__max_value:
            self.__value += 1
        else:
            print("Error: Exceeded the maximum value.")

    def reset(self):
        self.__value = 0

    def get_value(self):
        return self.__value

    def get_max_value(self):
        return self.__max_value

    def is_at_max(self):
        return self.__value == self.__max_value
```

9.Add a function in the base class of the object classes that returns true/false testing that the object overlaps with another object.

```python
class Shape:
    def area(self):
        raise NotImplementedError("area() not implemented")

    def perimeter(self):
        raise NotImplementedError("perimeter() not implemented")

    def get_x(self):
        raise NotImplementedError("get_x() not implemented")

    def get_y(self):
        raise NotImplementedError("get_y() not implemented")

    def boundary_points(self):
        raise NotImplementedError("boundary_points() not implemented")

    def contains_point(self, x, y):
        raise NotImplementedError("contains_point() not implemented")

    def overlaps(self, other):
        for (x, y) in self.boundary_points():
            if other.contains_point(x, y):
                return True

        for (x, y) in other.boundary_points():
            if self.contains_point(x, y):
                return True

        return False
```

11.Create a RasterDrawing class. Demonstrate that you can create a drawing made of several shapes, paint the drawing, modify the drawing, and paint it again.

```python
class RasterDrawing:
    def __init__(self):
        self.shapes = []

    def add_shape(self, shape):
        self.shapes.append(shape)

    def paint(self, canvas):
        canvas.clear_canvas()
        for shape in self.shapes:
            shape.paint(canvas)
```

```python
canvas = Canvas(40, 40)
drawing = RasterDrawing()

rect = Rectangle(10, 5, 5, 5, char='#')
circle = Circle(6, 20, 20, char='o')
triangle = Triangle(8, 6, 10, 25, char='^')

drawing.add_shape(rect)
drawing.add_shape(circle)
drawing.add_shape(triangle)

drawing.paint(canvas)
canvas.display()

rect.x = 15
rect.y = 2
circle.radius = 3

drawing.paint(canvas)
canvas.display()
```

12.Implement the ability to load/save raster drawings and demonstate that your method works. One way to implement this ability:
Overload __repr__ functions of all objects to return strings of the python code that would construct the object.

In the save method of raster drawing class, store the representations into the file.

Write a loader function that reads the file and uses eval to instantiate the object.

```python
import math

class Canvas:
    def __init__(self, width, height):
        self.width = width
        self.height = height
        self.data = [[' '] * width for _ in range(height)]

    def set_pixel(self, row, col, char='*'):
        if 0 <= row < self.height and 0 <= col < self.width:
            self.data[row][col] = char

    def clear_canvas(self):
        self.data = [[' '] * self.width for _ in range(self.height)]

    def display(self):
        print("\n".join("".join(row) for row in self.data))

class Shape:
    def paint(self, canvas):
        raise NotImplementedError


class Rectangle(Shape):
    def __init__(self, length, width, x, y, char='*'):
        self.length = length
        self.width = width
        self.x = x
        self.y = y
        self.char = char

    def boundary_points(self):
        points = []
        n = 4
        for i in range(n):
            points.append((self.x + i*self.length/n, self.y))
        for i in range(n):
            points.append((self.x + self.length, self.y + i*self.width/n))
        for i in range(n):
            points.append((self.x + self.length - i*self.length/n, self.y + self.width))
        for i in range(n):
            points.append((self.x, self.y + self.width - i*self.width/n))
        return points

    def __repr__(self):
        return f"Rectangle({self.length}, {self.width}, {self.x}, {self.y}, char={repr(self.char)})"

    def paint(self, canvas):
        for x, y in self.boundary_points():
            canvas.set_pixel(int(y), int(x), self.char)

class Circle(Shape):
    def __init__(self, radius, x, y, char='o'):
        self.radius = radius
        self.x = x
        self.y = y
        self.char = char

    def boundary_points(self):
        points = []
        n = 16
        for i in range(n):
            theta = 2 * math.pi * i / n
            points.append((
                self.x + self.radius * math.cos(theta),
                self.y + self.radius * math.sin(theta)
            ))
        return points

    def __repr__(self):
        return f"Circle({self.radius}, {self.x}, {self.y}, char={repr(self.char)})"

    def paint(self, canvas):
        for x, y in self.boundary_points():
            canvas.set_pixel(int(y), int(x), self.char)

class Triangle(Shape):
    def __init__(self, base, height, x, y, char='^'):
        self.base = base
        self.height = height
        self.x = x
        self.y = y
        self.char = char

    def boundary_points(self):
        points = []
        n = 5
        for i in range(n):
            points.append((self.x + i*self.base/n, self.y))
        for i in range(n):
            points.append((self.x, self.y + i*self.height/n))
        for i in range(n):
            t = i / n
            points.append((self.x + t*self.base, self.y + t*self.height))
        return points

    def __repr__(self):
        return f"Triangle({self.base}, {self.height}, {self.x}, {self.y}, char={repr(self.char)})"

    def paint(self, canvas):
        for x, y in self.boundary_points():
            canvas.set_pixel(int(y), int(x), self.char)


class RasterDrawing:
    def __init__(self):
        self.shapes = []

    def add_shape(self, shape):
        self.shapes.append(shape)

    def paint(self, canvas):
        canvas.clear_canvas()
        for shape in self.shapes:
            shape.paint(canvas)

    def save(self, filename):
        with open(filename, "w") as f:
            for shape in self.shapes:
                f.write(repr(shape) + "\n")


def load_drawing(filename):
    drawing = RasterDrawing()
    with open(filename, "r") as f:
        for line in f:
            drawing.add_shape(eval(line.strip()))
    return drawing
```

```python
canvas = Canvas(40, 40)

drawing = RasterDrawing()
drawing.add_shape(Rectangle(10, 5, 5, 5, char='#'))
drawing.add_shape(Circle(6, 20, 20, char='o'))
drawing.add_shape(Triangle(8, 6, 10, 25, char='^'))

drawing.paint(canvas)
canvas.display()

drawing.save("drawing.txt")

loaded = load_drawing("drawing.txt")
canvas2 = Canvas(40, 40)
loaded.paint(canvas2)
canvas2.display()
```
